```python
import torch
import torch.nn.functional as F
import matplotlib.pyplot as plt
```

```python
# no nn.Linear, just raw tensors
x = torch.randn(4, 8)  # batch=4, dim=8
W = torch.randn(8, 16, requires_grad=True)
b = torch.randn(16, requires_grad=True)

out = x @ W + b
print(out.shape)  # (4, 16)
```

```python
loss = out.sum()
loss.backward()
print(W.grad.shape)  # should be (8, 16)
print(b.grad.shape)  # should be (16,)
```

```python
def softmax(x: torch.Tensor) -> torch.Tensor:
    e = torch.exp(x - x.max(dim=-1, keepdim=True).values)  # stable
    return e / e.sum(dim=-1, keepdim=True)

# sanity check
logits = torch.randn(4, 8)
probs = softmax(logits)
print(probs.sum(dim=-1))  # should be all 1.0
```

# Attention

```python
def attention(
    Q: torch.Tensor,  # (batch, seq, d_k)
    K: torch.Tensor,  # (batch, seq, d_k)
    V: torch.Tensor,  # (batch, seq, d_v)
) -> torch.Tensor:   # (batch, seq, d_v)
    d_k = Q.shape[-1]
    scores = Q @ K.transpose(-2, -1) / d_k**0.5
    weights = softmax(scores)  # use your softmax from above
    return weights @ V
```

```python
B, S, D = 2, 6, 32
Q = torch.randn(B, S, D)
K = torch.randn(B, S, D)
V = torch.randn(B, S, D)

out = attention(Q, K, V)
print(out.shape)  # (2, 6, 32)
```
